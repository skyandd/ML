```python
import numpy as np
from sklearn.datasets import load_iris
from MLalgorithms.pca import PCA
import matplotlib.pyplot as plt
%matplotlib inline
```

Математика работы алгоритма представлена в статье https://habr.com/ru/post/304214/

Основной же алгоритм работы следущий

- Центрируем данные $$X = X - \mu$$
- Находим матрицу ковариации наших признаков  $$cov = cov(X.T)$$
- Находим собственнные значение и вектора нашей ковариационной матрицы $$\nu, \lambda = eig(cov)$$
- Сортируем по убыванию собственные значения
- Выбираем первые N(количество главных компонент) собственных значений и их соответсвующие вектора
- Матрично умножаем наши центрированные данные на N выбранных собственных векторов $$\tilde{X}= X(\nu .T)$$

```python
data = load_iris()
X = data.data
y = data.target
```

```python
pca = PCA(2)
```

```python
pca.fit(X)
```

```python
X_transform = pca.transform(X)
```

```python
plt.scatter(X_transform[:, 0], X_transform[:, 1], c = y)
```
